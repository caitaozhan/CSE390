```python
import numpy as np
import scipy.special as sps
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def draw_gamma(alpha, beta):
    S = np.random.gamma(alpha, 1.0/beta, size=100000)
    series = pd.Series(S)
    plt.title("Gamma(%1.4f, %1.4f), mean=%1.4f, var=%1.4f" % (alpha, beta, series.mean(), series.std()*series.std()), fontsize=16)
    series.plot.hist(bins=200, normed=True, figsize=(10,7))
```

## mean=1, var=1

```python
alpha = 1.0        # shape
beta  = 1.0        # rate
```

```python
draw_gamma(alpha, beta)
```
